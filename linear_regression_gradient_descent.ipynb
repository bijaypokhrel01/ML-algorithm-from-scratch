**Importing all the dependencies**

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

**Now, I will create a gradient descent function which returns a new weight and bias**

```python
def gradient_descent(m, b, X, y, alpha):
    grad_m = 0
    grad_b = 0

    for i in range(len(X)):
        grad_m += -1/(len(X)) * (y[i] - (m * X[i, 0] + b)) * X[i, 0]
        grad_b += -1/(len(X)) * (y[i] - (m * X[i, 0] + b))
    
    new_m = m - alpha * grad_m
    new_b = b - alpha * grad_b
    
    return new_m, new_b
```

##### I used sklearn to create a regression datasets

```python
import sklearn.datasets as data

X, y = data.make_regression(n_samples=100, n_features=1, noise=20, random_state=4)
len(X), len(y)
```

##### Lets visualize how our data is spreadout using scatter plot

```python
plt.scatter(X[:, 0], y, color='black')
plt.show()
```

* **Now initialize weight and bias**<br></br>
* **Choose appropriate learning rate and epoches so that the model could able to find the best fit predicted line.**

```python
w = 0
b = 0
alpha = 1.07
epoches = 100


for j in range(epoches):
    if j % 50 == 0:
        print(f"Epoches: {j}")

    m, b = gradient_descent(w, b, X, y, alpha)
```

**I choose and check wether the line fits our model. And with lot of hit and trial method finally i came with the appropriate learning rate alpha = 1.07**

```python
plt.scatter(X, y, color='black', label='Actual data')
plt.plot(X[:, 0], [(m * x + b) for x in X[:, 0]], color='red', label='Predicted line')
plt.legend()
plt.xlabel('Features variable')
plt.ylabel('Target variable')
plt.title('Regression model from scratch')
plt.show()
```

```python
prediction = m * 2.1 + b
print(f"The prediction of salary for 3.4 years of work experience is {prediction:.2f}")
```
